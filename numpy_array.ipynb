```python
### creatting array manually

a = np.array([0,1,2,3,4,5])                     #vector is one dimensional array

print(type(a))

###areating an array not manually

a = np.arange(6)

print(type(a))
```

```python
## creating 2d array

two_d = np.array([[1,2,3],[1,2,3]])                 #matrix is two dimensional array
print(two_d)
print(len(two_d))
print(np.shape(two_d))
print(two_d.ndim)
```

```python
### creating 3d array

three_d = np.array( [ [ [1,2,3] , [1,5,3] ] , [ [1,2,3] , [1,2,3] ] ] )

print(three_d)

print(three_d.ndim)

print(three_d.shape)

print(len(three_d))
```

```python
### slicing of numpy array

a = np.array([1,2,3,4,5])

b = a[ : :-1]

b[0]=10
print(a)
print(b)

np.may_share_memory(a,b)                         #tells weather two variables point to same location or not
```

```python

a = np.array([1,2,3,4,5])

b = a[ : :-1].copy()

b[0]=10
print(a)
print(b)

np.may_share_memory(a,b)
```

```python
#### fancy indexing

a = np.random.randint(0,20,15)

print(a)
mask = (a%2 == 0)
b = a[mask]
print(b)
```

```python
## assigning values through mask

print(a)
a[mask]=-5

print(a)
```

```python
#### indexing numpy with integer arrays

c = a[[1,2,3,4]]
c
```
