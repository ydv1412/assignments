```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
plt.xlabel("values")
plt.ylabel("index")
plt.title("graph")
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
t = np.arange(0 ,5.,0.2)
plt.plot(t,t**2,'r--',label='^2')
plt.plot(t,t**2.2,'b^',label='^2.2')
plt.plot(t,t**2.5,'gs',label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
### controling line width usinf keyword arguments

plt.plot([1,2,3,4],linewidth="5.")
plt.show()
```

```python
x1 = [1,2,3,4]
y1 = [1,4,9,16]
x2 = [1,2,3,4]
y2 = [2,4,6,8]

line = plt.plot(x1,y1,x2,y2)

plt.setp(line[0],color='r',linewidth='5')   ###in python we can format by keyword arguments
plt.setp(line[1],'color','g','linewidth',2)   ###also supports matlab type arguments
plt.grid()
plt.show()
```

```python
def f1(t):
    return np.exp(-t)* np.cos(2*np.pi*t)
def f2(t):
    return np.exp(t)* np.cos(2*np.pi*t)
def f3(t):
    return np.exp(-t)* np.sin(2*np.pi*t)
def f4(t):
    return np.exp(t)* np.sin(2*np.pi*t)    

t = np.arange(0,1,0.1)
plt.figure(1)
plt.subplot(221)
plt.plot(t,f1(t),color='b')
plt.title("first")

plt.subplot(222)
plt.plot(t,f2(t),color='g')
plt.title("second")

plt.subplot(223)
plt.plot(t,f3(t),color='r')
plt.title("third")

plt.subplot(224)
plt.plot(t,f4(t),color='y')
plt.title("fourth")

plt.show()
```
