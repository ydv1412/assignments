```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

x = stats.norm.rvs(size=1000)   ### creating a standard gaussian distribution
sns.set_style('whitegrid')
sns.kdeplot(np.array(x),bw=0.5)        #### kdeplot function is used for pdf and bw parameter is used as smoothing parameter
print(stats.kstest(x,'norm'))                ### kstest is a function which returns test stats and p-value respectively
plt.plot()
```

here we can see that the test value is much less than p-value hence both the variables are similar

```python
### for different distribution


x = np.random.uniform(0,1,1000)
sns.kdeplot(np.array(x),bw=0.1)
print(stats.kstest(x,'norm'))
plt.show()
```
